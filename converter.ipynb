3) read as tsv

```python
import csv
```

```python
# input_name = "../notes/snk-vocab-series.tsv"
input_name = "../notes/snk-vocab-series-n1-only.tsv"
output_name = "../parsed/anki-deck-snk-vocab-n1-only.txt"
```

```python
with open(input_name, 'r') as file:
    tsv_reader = csv.reader(file, delimiter='\t')
    data = list(tsv_reader)
```

```python
print(len(data))
print(data[0])
```

```python
word_list = list(item[0] for item in data)
print(len(word_list))
print(word_list[0:5])
```
