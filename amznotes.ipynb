```python
# FILE NAME
# FILE_NAME = "../notes/LN-AMZ-notes-妹はカノジョにできないのに ２ (電撃文庫)-Notebook.html"
# FILE_NAME = "../notes/LN-AMZ-notes-幼なじみからの恋愛相談。2　相手は俺っぽいけど違うらしい (角川スニーカー文庫) (Japanese Edition)-Notebook.html"
FILE_NAME = "../notes/LN-AMZ-notes-幼なじみからの恋愛相談。　相手は俺っぽいけど違うらしい (角川スニーカー文庫) (Japanese Edition)-Notebook.html"
OUTPUT_FILE = "../parsed/highlights-renaisoudan-1.txt"
```

```python
# LOAD HTML FILE
file_handler = open(FILE_NAME, "r", encoding="utf-8")
html = file_handler.read()
soup = BeautifulSoup(html, 'html.parser')
```

```python
# EXTRACT HIGHLIGHTS
highlights = []
for div in soup.findAll("div",{"class":"noteText"}):
    highlights.append(div.contents[0])
print(highlights)
```

```python
# REMOVE SPACES, and all punctuation
new_hl = []
punc = ''' !()-[]{};:'"\,<>./?@#$^&*_~「」（）・＿。｛｝；：！
『』【】…？,、－、
'''
for hl in highlights:
    for c in hl:
        if c in punc:
            hl = hl.replace(c, "")

    new_hl.append(hl.replace(" ", ""))
print(new_hl)
```

```python
# REMOVE DUPLICATES
print(len(new_hl))
new_hl = set(new_hl)
print(len(new_hl))
```

```python
# SORTED

# TOOK FROM CHATGPT IDK IF IT WORKS???
# Define the custom sorting key
def japanese_sort_key(word):
    # Convert the word to hiragana using the mojimoji library
    import mojimoji
    hiragana_word = mojimoji.han_to_zen(word, digit=False, ascii=False, kana=True)
    
    # Split the word into individual characters
    characters = [c for c in hiragana_word]
    
    # Convert each character to its "gojuuon" order
    gojuuon_order = []
    for c in characters:
        if c in ('あ', 'か', 'さ', 'た', 'な', 'は', 'ま', 'や', 'ら', 'わ'):
            gojuuon_order.append(c)
        elif c in ('い', 'き', 'し', 'ち', 'に', 'ひ', 'み', 'り'):
            gojuuon_order.append(c + '1')
        elif c in ('う', 'く', 'す', 'つ', 'ぬ', 'ふ', 'む', 'ゆ', 'る'):
            gojuuon_order.append(c + '2')
        elif c in ('え', 'け', 'せ', 'て', 'ね', 'へ', 'め', 'れ'):
            gojuuon_order.append(c + '3')
        elif c in ('お', 'こ', 'そ', 'と', 'の', 'ほ', 'も', 'よ', 'ろ', 'を'):
            gojuuon_order.append(c + '4')
        else:
            gojuuon_order.append(c)
    
    # Join the "gojuuon" order characters back into a string
    gojuuon_word = ''.join(gojuuon_order)
    
    # Return the "gojuuon" order string as the sorting key
    return gojuuon_word

```

```python
# Sort the list of Japanese words using the custom sorting key
sorted_hl = sorted(new_hl, key=japanese_sort_key)
print(sorted_hl)
```
